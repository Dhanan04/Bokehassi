```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
#Q1
from bokeh.plotting import figure, output_file , show
```

```python
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
#sample data set of plot on Bokeh
output_file('flower.html')
p = figure(title = 'flower data')
p.xaxis.axis_label = 'sepal_length'
p.yaxis.axis_label = 'sepal_width'
p.circle(flowers ['sepal_length'] , flowers['sepal_width'])
show(p)
```

 glyphs are the visual markers used to represent data in plots they are lines, circle, triangle ,rectangel and square and they are added with figure

```python
x = [4,5,6,7,8,9]
y = [5,6,7,8,2,3]
output_file('test.html')
p = figure(title = 'Test data')
p.square( x , y  ,fill_color = 'green', size = 30)
show(p)
```

```python
x = [4,5,6,7,8,9]
y = [5,6,7,8,2,3]
output_file('test.html')
p = figure(title = 'Test data')
p.triangle( x , y  ,fill_color = 'green' , size = 30)
show(p)
```
